```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#the total number of schools
total_school = len(school_data_complete["school_name"].unique())
#the total number of students
total_student = len(school_data_complete["Student ID"].unique())
#the total budget
total_budget = school_data["budget"].sum()
#the average maths score
average_maths = school_data_complete["maths_score"].mean()
#the average reading score
average_reading = school_data_complete["reading_score"].mean()
#Total number of students for math 
total=school_data_complete["maths_score"].count()
# % passing math
pass_maths_df = school_data_complete.loc[(school_data_complete["maths_score"]>=50), ["maths_score"]]
passmaths=pass_maths_df["maths_score"].count()
pcpassmaths=passmaths/total*100
# % passing reading
passreading_df = school_data_complete.loc[(school_data_complete["reading_score"]>=50), ["reading_score"]]
passreading=passreading_df["reading_score"].count()
pcpassreading=passreading/total*100
# %
pass_maths_reading_df=school_data_complete.loc[(school_data_complete["maths_score"]>=50)&(school_data_complete["reading_score"]>=50),:]
pcOverallPassing=pass_maths_reading_df["reading_score"].count()/total*100

#Create a dataframe to hold the above results
area_summary = pd.DataFrame({"Total Schools":[total_school],
                                    "Total Students":[total_student],
                                    "Total Budget":[total_budget],
                                    "Average Maths Score":[average_maths],
                                    "Average Reading Score":[average_reading],
                                    "% Passing Maths":[pcpassmaths],
                                    "% Passing Reading":[pcpassreading],
                                    "% Overall Passing":[pcOverallPassing]})

#Optional: give the displayed data cleaner formatting
area_summary.loc[:,"Total Budget"]="$" + area_summary["Total Budget"].map('{:,.2f}'.format)
area_summary.loc[:,"Total Students"]=area_summary["Total Students"].map('{:,.0f}'.format)

area_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
df1=school_data_complete.groupby(["school_name","type"])

budget_by_school=school_data_complete.groupby("school_name").mean()["budget"].reset_index()

df2=df1.count().reset_index()
df3=df2[["school_name","type","Student ID"]]

df4=df3.loc[:,["school_name","type","Student ID"]]
df4["budget"]=budget_by_school["budget"]
df4["Per Student Budget"]=df4["budget"]/df4["Student ID"]

#average maths and reading score
dfsum=df1.sum().reset_index()
df4["Average Maths Score"]=dfsum["maths_score"]/df4["Student ID"]
df4["Average Reading Score"]=dfsum["reading_score"]/df4["Student ID"]

#% passing for maths
df_passmaths_byschool=school_data_complete.groupby('school_name')['maths_score'].apply(lambda x: (x>=50).sum()).reset_index()
df4["% Passing Maths"]=df_passmaths_byschool["maths_score"]/df4["Student ID"]*100

#% passing for reading
df_passreading_byschool=school_data_complete.groupby('school_name')['reading_score'].apply(lambda x: (x>=50).sum()).reset_index()
df4["% Passing Reading"]=df_passreading_byschool["reading_score"]/df4["Student ID"]*100

#Use df_passboth_byschool from previous cell before grouping to calculate % Overall Passing
df_passboth_byschool=pass_maths_reading_df.groupby('school_name').count().reset_index()
df4["% Overall Passing"]=df_passboth_byschool["Student ID"]/df4["Student ID"]*100

#Final formatting:
df5=df4.rename(columns={"school_name":"","type":"School Type","Student ID":"Total Students","budget":"Total School Budget"})
df5.loc[:,"Per Student Budget"]="$"+df5["Per Student Budget"].map('{:,.2f}'.format) #format to currency
df5.loc[:,"Total School Budget"]="$" + df5["Total School Budget"].map('{:,.2f}'.format) #format to currency

per_school_summary=df5.set_index("")
per_school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
#Use df4 in first cell, before formatting so that Per Student Budget is still float64

df4["Spending Ranges (Per Student)"]=pd.cut(df4["Per Student Budget"],spending_bins,labels=labels)
school_spending_df=df4.groupby(["Spending Ranges (Per Student)"])
spending_maths_scores = school_spending_df["Average Maths Score"].mean()
spending_reading_scores = school_spending_df["Average Reading Score"].mean()
spending_passing_maths = school_spending_df["% Passing Maths"].mean()
spending_passing_reading = school_spending_df["% Passing Reading"].mean()
overall_passing_spending = school_spending_df["% Overall Passing"].mean()
spending_summary=pd.DataFrame({'Average Maths Score':spending_maths_scores,
                             'Average Reading Score':spending_reading_scores,
                             '% Passing Maths':spending_passing_maths,
                             '% Passing Reading':spending_passing_reading,
                             '% Overall Passing':overall_passing_spending})
spending_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
#Use df4 in first cell, where school size or number of total students = "Student ID"
df4["School Size"]=pd.cut(df4["Student ID"],size_bins,labels=labels)
group_by_school_size=df4.groupby(["School Size"])
size_maths_scores=group_by_school_size.mean()["Average Maths Score"]
size_reading_scores=group_by_school_size.mean()["Average Reading Score"]
size_passing_maths=group_by_school_size.mean()["% Passing Maths"]
size_passing_reading=group_by_school_size.mean()["% Passing Reading"]
size_overall_passing=group_by_school_size.mean()["% Overall Passing"]

size_summary=pd.DataFrame({'Average Maths Score':size_maths_scores,
                           'Average Reading Score':size_reading_scores,
                             '% Passing Maths':size_passing_maths,
                             '% Passing Reading':size_passing_reading,
                             '% Overall Passing':size_overall_passing})
size_summary
```
